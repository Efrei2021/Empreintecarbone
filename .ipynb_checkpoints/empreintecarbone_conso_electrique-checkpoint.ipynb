```python
PATH_DIRECTORY = os.getcwd()
PATH_DATA = os.path.join(PATH_DIRECTORY, "Donneses")
PATH_DATA_PUISSANCE = os.path.join(PATH_DATA, "mes-puissances-atteintes-30min-004043845308-94550.csv")
PATH_DATA_TAUX = os.path.join(PATH_DATA, r"eCO2mix_RTE_2023-12-23.xlsx")
```

# Consommation électrique

```python
df_puissance =pd.read_csv(PATH_DATA_PUISSANCE, sep=";", encoding="latin-1")
df_puissance
```

```python
# Structuration de df_puissance
df_puissance_structured=df_puissance.copy()
df_puissance_structured.reset_index(inplace=True)

# Change les colonnes
df_puissance_structured.rename(columns={'level_0':'Date et heure de relève par le distributeur',
                                        'level_1':'Puissance atteinte (W)',
                                        'Récapitulatif de mes puissances atteintes en W':'Nature de la donnée'},inplace=True)
df_puissance_structured=df_puissance_structured.iloc[1:, :]
df_puissance_structured
```

```python
# Information sur la base de données
df_puissance_structured.info()
```

```python
index_jour = df_puissance_structured[df_puissance_structured["Date et heure de relève par le distributeur"]=="23/12/2023"].index[0]

df_puissance_jourj = df_puissance_structured.iloc[index_jour:index_jour+48, :]

# Création de la colonne Date+heure
df_puissance_jourj["Datetime"] = [pd.to_datetime("23/12/2023"+" "+time, format="%d/%m/%Y %H:%M:%S") for time in df_puissance_jourj["Date et heure de relève par le distributeur"].to_list()]
df_puissance_jourj = df_puissance_jourj.sort_values(by="Datetime")
df_puissance_jourj.drop([df_puissance_jourj.columns[0], df_puissance_jourj.columns[2]], axis=1, inplace=True)

#Concertir les données de puissance en entier
df_puissance_jourj['Puissance atteinte (W)'] = df_puissance_jourj['Puissance atteinte (W)'].astype("int")

#Indexer les données 
df_puissance_jourj.set_index('Datetime', inplace=True)

#Calcul de la consommation électrique
df_conso_elec = df_puissance_jourj.copy()
df_conso_elec["Consommation électrique atteinte (kWh)"] = [round((puissance/2000),1) for puissance in df_conso_elec['Puissance atteinte (W)']]

df_conso_elec
```

# Taux de CO2

```python
df_eco2mix=pd.read_excel(PATH_DATA_TAUX)
df_eco2mix = df_eco2mix[['Date','Heures','Taux de Co2']].iloc[:-1, :]
df_eco2mix
```

```python
# Information sur la base de données
df_eco2mix.info()
```

```python
# Création de la colonne Date+heure
df_eco2mix_structured = df_eco2mix.copy()
df_eco2mix_structured['Datetime'] = [datetime.datetime.combine(date.date(),time) for date, time in zip(df_eco2mix_structured["Date"].to_list(), df_eco2mix_structured["Heures"].to_list())]
df_eco2mix_structured = df_eco2mix_structured.sort_values(by="Datetime")
df_eco2mix_structured.drop([df_eco2mix_structured.columns[0], df_eco2mix_structured.columns[1]], axis=1, inplace=True)

#Indexer les données 
df_eco2mix_structured.set_index('Datetime', inplace=True) 
df_eco2mix_structured.rename(columns={'Taux de Co2':'Taux de Co2 (gCO2eq/kWh)'},inplace=True)

df_eco2mix_structured
```

```python
# Ré-échantillonnage df eco2mix
df_eco2mix_resample = df_eco2mix_structured.copy()
df_eco2mix_resample=df_eco2mix_resample.resample('30T').mean()
df_eco2mix_resample
```

```python
df_eco2mix_resample.plot()
```

# Aggregation des données

```python
df_agg = pd.concat([df_conso_elec, df_eco2mix_resample], axis=1)
df_agg['empreintecarbone_tempsréel (gCO2eq)'] = [(c*t) for c, t in zip(df_agg[df_agg.columns[1]].to_list(), df_agg[df_agg.columns[2]].to_list())]
df_agg.head()
```
